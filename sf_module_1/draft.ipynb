```python
data['length_of_title'] = data.original_title.apply(lambda s: len(s))

from collections import defaultdict
from statistics import mean

d = defaultdict(list)

for _, row in data.iterrows():
    for c in row.production_companies.split('|'):
        d[c].append(row.length_of_title)
        
series = pd.Series([mean(i) for i in d.values()], index=d.keys())
display(series.index[series.max()])

data.pivot_table(
    values='length_of_title',
    index='production_companies',
    aggfunc='mean'
).query('length_of_title == length_of_title.max()')



```
